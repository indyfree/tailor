```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import os

my_path = os.path.abspath(os.path.dirname("__file__"))
path_iris = os.path.join(my_path, '..', "data/test/iris.csv")
path_iris = os.path.normpath(path_iris)
print(path_iris)
df_iris = pd.read_csv(path_iris)
```

```python
print(df_iris)
```

```python
plt.scatter(df_iris['species'], df_iris['sepal_length'])
#referencing certain columns in DataFrame is only possible with the column label
```

```python
plt.scatter(x=df_iris['petal_length'], y=df_iris['petal_width'])
```

```python
#ugly af
color_species = ['red' if s == 'setosa' else ('blue' if s == 'versicolor' else 'green') for s in df_iris['species']]
print(color_species)
```

```python
#easier on the eye
def color_for_species(species):
    if species == 'setosa':
        return 'red'
    elif species == 'versicolor':
        return 'blue'
    else:
        return 'green'
    
color_species = [color_for_species(s) for s in df_iris['species']]
print(color_species)
```

```python
#using dictionary
def color_for_species_dict(species):
    return {
        'setosa':'red',
        'versicolor': 'blue',
        'virginica' : 'green'
    }.get(species, 'black') #black is default if not found

color_species = [color_for_species_dict(s) for s in df_iris['species']]
print(color_species)
```

```python
# using pyplot plotting
plt.scatter(x=df_iris['petal_length'], y=df_iris['petal_width'], c=color_species)
```

```python
# using DataFrame plotting
df_iris.plot(kind='scatter', x='petal_length', y='petal_width', c=color_species)
```
